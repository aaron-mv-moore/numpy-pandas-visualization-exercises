```python
import matplotlib.pyplot as plt
import numpy as np


# Set retina for higher resolution display - specific for jupyter notebook
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

# Subplots

```python
x = range(1, 10)
y1 = [1 for n in x]
y2 = [np.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * np.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [np.math.factorial(n) for n in x]
y8 = [n**n for n in x]


fig, ax = plt.subplots()
plt.suptitle('Big O Notation')

plt.subplot(241)
plt.plot(x, y1 , c='r')
plt.title('$O(1)$')

plt.subplot(242)
plt.plot(x, y2 , c='orange')
plt.title('$O(log n)$')

plt.subplot(243)
plt.plot(x, y3, c='green')
plt.title('$0(n)$')

plt.subplot(244)
plt.plot(x, y4, c='blue')
plt.title('$O(n log n)$')

plt.subplot(245)
plt.plot(x, y5, c='indigo')
plt.title('$O(n^2)$')

plt.subplot(246)
plt.plot(x, y6, c='violet')
plt.title('$O(2^n)$')

plt.subplot(247)
plt.plot(x, y7, c='teal')
plt.title('$O(n!)$')

plt.subplot(248)
plt.plot(x, y7, c='yellow')
plt.title('$O(n^n)$')


fig.supxlabel('Operations')
fig.supylabel('Elements')
plt.tight_layout()
plt.show()
```
