5. Percentage of Pokemon whose Type 2 is Ghost

    - Create a function named percent_ghost
    - Use the loaded dataframe as an argument

```python
def percent_ghost(df):
    percent_ghost = round(sum(df['Type 2'] == 'Ghost')/df.shape[0] * 100, 2)
    return print(f'{percent_ghost}% of the pokemon in this dataset have Ghost Type 2.')

percent_ghost(df)
```

```python
# List method
# making the columns a list for easier manipulation
column_name = list(df.columns)
column_name

# testing lower lsit comp for correctness
column_name = [i.lower() for i in column_name]

# replacing the the '.' with nothing
column_name = [i.replace('.', '') for i in column_name]

# stripping trailing spaces
column_name = [i.strip() for i in column_name]

# changing spaces to underscores
column_name = [i.replace(' ','_') for i in column_name]
column_name
```
