### 1. Use matplotlib to plot the following equation:

    - y=x^2âˆ’x+2

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.

```python
x = range(1, 101)
y = [n**2 - n + 2 for n in x]

plt.annotate('Origin', xy=(0, 0), xytext=(5, 2000),
             arrowprops={'facecolor': 'red'})

plt.plot(x, y)
plt.show()
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python
x = range(1, 101)
y1 = [np.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n +1) for n in x]


plt.suptitle('Combo Deal')

plt.subplot(221)
plt.plot(x, y1, c = 'r')

plt.subplot(222)
plt.plot(x, y2, c = 'orange')

plt.subplot(223)
plt.plot(x, y3, c = 'yellow')

plt.subplot(224)
plt.plot(x, y4, c = 'g')

plt.show()
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = range(1, 5)
y1 = [np.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n +1) for n in x]
plt.plot(x, y1, c='g', label = '$y=\sqrt{x}$')
plt.plot(x, y2, c='red', label='$y=x^3$')
plt.plot(x, y3, c='gray', label='$y=2^x$')
plt.plot(x, y4, c='y', label='$y=1/(x+1)}$')

plt.title('Lines')
plt.legend(loc=2)
plt.show()
```
